```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
#ECM

edml_dc_ecm = pd.read_csv('Specific Sulfate/EDML/EDML ECM.txt', comment='#', sep='\t', header=0, names=['depth', 'dc_ecm']) 

edml_dc_ecm.sort_values('depth', inplace=True)

edml_dc_ecm.iloc[:,1] = edml_dc_ecm.iloc[:,1].mask(edml_dc_ecm.iloc[:,1] < 0)
edml_dc_ecm.to_csv('lc_data_for_matchmaker/edml/edml_dc_ecm.txt', index=False)
```

```python
edml_sulfate = pd.read_csv('Specific Sulfate/EDML/Severi EDML Sulfate.txt', sep=r'\s+', skiprows=3, names=['depth', 'sulfate']) 

edml_sulfate.sort_values('depth', inplace=True)


edml_sulfate.iloc[:,1] = edml_sulfate.iloc[:,1].mask(edml_sulfate.iloc[:,1] < 0)
edml_sulfate.to_csv('lc_data_for_matchmaker/edml/edml_sulfate.txt', index=False)
```

```python
#LC Things

lc_path1='Specific Sulfate/EDML/EDML_CFA_113-1443.499m_1mm_resolution.txt'
lc_path2='Specific Sulfate/EDML/EDML_CFA_1443.5-2774m_1mm_resolution.txt'

read_lc1 = pd.read_csv(lc_path1, header=None, skiprows=1, sep='\t', names=['Depth(m)', 'Na(ng/g)','NH4(ng/g)', 'Ca(ng/g)', 'Dust(particles/ml)', 'Cond(mikroS/cm)'])
read_lc2 = pd.read_csv(lc_path2, header=None, skiprows=1, sep='\t', names=['Depth(m)', 'Na(ng/g)','NH4(ng/g)', 'Ca(ng/g)', 'Dust(particles/ml)', 'Cond(mikroS/cm)'])

read_lc = pd.concat([read_lc1, read_lc2], ignore_index=True)

name = 'nh4'
save = read_lc[["Depth(m)", "NH4(ng/g)"]].copy(deep=True)
save.iloc[:,1] = save.iloc[:,1].mask(save.iloc[:,1] < 0)
save.to_csv(f'lc_data_for_matchmaker/edml/edml_{name}.txt', index=False)

name = 'ca'
save = read_lc[["Depth(m)", "Ca(ng/g)"]].copy(deep=True)
save.iloc[:,1] = save.iloc[:,1].mask(save.iloc[:,1] < 0)
save.to_csv(f'lc_data_for_matchmaker/edml/edml_{name}.txt', index=False)

name = 'dust'
save = read_lc[["Depth(m)", "Dust(particles/ml)"]].copy(deep=True)
save.iloc[:,1] = save.iloc[:,1].mask(save.iloc[:,1] < 0)
save.to_csv(f'lc_data_for_matchmaker/edml/edml_{name}.txt', index=False)

name = 'na'
save = read_lc[["Depth(m)", "Na(ng/g)"]].copy(deep=True)
save.iloc[:,1] = save.iloc[:,1].mask(save.iloc[:,1] < 0)
save.to_csv(f'lc_data_for_matchmaker/edml/edml_{name}.txt', index=False)

name = 'Cond'
save = read_lc[["Depth(m)", "Cond(mikroS/cm)"]].copy(deep=True)
save.iloc[:,1] = save.iloc[:,1].mask(save.iloc[:,1] < 0)
save.to_csv(f'lc_data_for_matchmaker/edml/edml_{name}.txt', index=False)
```
