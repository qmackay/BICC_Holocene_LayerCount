```python
import pandas as pd
import os
import numpy as np
os.chdir('/Users/quinnmackay/Documents/GitHub/BICC_Holocene_LayerCount/data/raw_lcs')
```

```python
#EDML First

revised_count = pd.read_csv('Modified LCs/EDML_LayerCount_Revised.txt', sep='\t')
revised_count = revised_count[['Depth (m)', 'count', 'age (b1950)']]

original_count = pd.read_csv('Original Raw LCs/edml_lc.txt', sep='\t')
original_count = original_count[['Depth', 'Count']]
original_count = original_count[original_count['Depth'] > np.max(revised_count['Depth (m)'])] #only keep the part of the original count that is deeper than the revised count

output_count = pd.concat([revised_count, original_count.rename(columns={'Depth': 'Depth (m)', 'Count': 'count'})], ignore_index=True)
output_count.sort_values('Depth (m)', inplace=True)

for _, row in output_count.iterrows():

    if row['age (b1950)'] > 0:
        lastval = row['age (b1950)']
        continue

    else:
        row['age (b1950)'] = lastval+row['count']
        lastval = row['age (b1950)']

output_count.rename(columns={'age (b1950)': 'Age', 'Depth (m)': 'Depth', 'count': 'Count'}, inplace=True)

os.makedirs('Counts for Matchmaker', exist_ok=True)

output_count.to_csv('Counts for Matchmaker/EDML_LayerCount.txt', sep='\t', index=False)
```

```python
#Now WDC

revised_count = pd.read_csv('Modified LCs/Updated_WD2014 Layer Count.tab', sep='\t', names=['Depth', 'Age'], skiprows=1)
revised_count['Age'] = revised_count['Age'] * 1000 #convert from ka to years

output_count = revised_count

output_count.to_csv('Counts for Matchmaker/WDC_LayerCount.txt', sep='\t', index=False)
```
